```python
import numpy as np
from scipy.stats import truncnorm
import random
import math
import copy
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import matplotlib.cm as cm
import matplotlib


def generate_data(num_of_data):
    input_data=[]
    output_data=[]
    current_input_value=0
    current_input_value_temp=0
    current_output_value=0
    for i in range(num_of_data):
        if (random.random()<0.1):
            current_input_value_temp=random.random()
        current_input_value=current_input_value+(current_input_value_temp-current_input_value)*(random.random())/2
        current_output_value=current_input_value            
        input_data.append(current_input_value)
        output_data.append(current_output_value)
    return input_data,output_data

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (20, 6)

input_data,output_data=generate_data(500)
plt.plot(input_data)
plt.plot(output_data)
plt.show()
```
